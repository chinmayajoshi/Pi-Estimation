```python
import random 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_random():
    v = random.random()
    sign = random.randint(0, 1)
    if not sign: v *= -1
    return v

def is_in_circle(x, y):
    return (x**2 + y**2) < 1
    
def cal_pi(N):
    c = 0
    xs = []
    ys = []
    is_circle = []
    for _ in range(N):
        x = get_random()
        y = get_random()
        in_circle = is_in_circle(x, y)
        if in_circle: c += 1

        xs.append(x)
        ys.append(y)
        is_circle.append(in_circle) 

    pi =  4* c / N
    return pi, [xs, ys, is_circle]
```

```python
N = 1_000_000
pi, pts = cal_pi(N)
print(pi)
```

```python
sns.scatterplot(x=pts[0], y=pts[1], hue=pts[2])
plt.legend(loc=(1, 0.5))
```
